```python
import os
import json
import librosa
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
#오디오 데이터셋 초기화
class AudioDataset(Dataset):
    def __init__(self, root_audio_dir, root_label_dir, fmin='C2', n_bins=60, bins_per_octave=12, higher_resolution=False):
        self.root_audio_dir = root_audio_dir
        self.root_label_dir = root_label_dir
        self.file_names = []
        self.raw_labels = []  # 원래의 문자열 라벨을 저장하는 리스트
        self.fmin = librosa.note_to_hz(fmin)
        self.n_bins = n_bins
        self.bins_per_octave = bins_per_octave
        if higher_resolution:
            self.n_bins *= 2
            self.bins_per_octave *= 2
        self.load_files_and_labels()

        # LabelEncoder로 라벨 인코딩
        self.label_encoder = LabelEncoder()
        self.labels = self.label_encoder.fit_transform(self.raw_labels)
        
    #데이터와 레이블 로드
    def load_files_and_labels(self):
        for dirpath, _, filenames in os.walk(self.root_audio_dir):
            for filename in filenames:
                if filename.endswith('.wav'):
                    full_path = os.path.join(dirpath, filename)
                    self.file_names.append(full_path)
                    label_path = os.path.join(self.root_label_dir, filename.replace('.wav', '.json'))
                    label = self.load_label(label_path)
                    self.raw_labels.append(label)
    #레이블 파일 로드
    def load_label(self, label_path):
        try:
            if os.path.exists(label_path):
                with open(label_path, 'r', encoding='utf-8') as json_file:
                    data = json.load(json_file)
                    return data['전사정보']['LabelText']
        except json.JSONDecodeError as e:
            print(f"Error reading {label_path}: {e}")
        return "No Label Found"


    def __getitem__(self, idx):
        file_path = self.file_names[idx]
        label = self.labels[idx]  # 인덱스를 사용하여 라벨 배열에서 라벨 검색

        y, sr = librosa.load(file_path, sr=None)
        C = np.abs(librosa.cqt(y, sr=sr, fmin=self.fmin, n_bins=self.n_bins, bins_per_octave=self.bins_per_octave))
        C_db = librosa.amplitude_to_db(C, ref=np.max)

        return torch.tensor(C_db, dtype=torch.float32), label

    def __len__(self):
        return len(self.file_names)
```

```python
class PitchCNN(nn.Module):
    def __init__(self, num_classes):
        super(PitchCNN, self).__init__()
        
        self.layer1 = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2))
        
        self.layer2 = nn.Sequential(
            nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2))
        
        self.layer3 = nn.Sequential(
            nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2))
        
        # 적응형 평균 풀링 층을 추가
        self.adaptive_pool = nn.AdaptiveAvgPool2d((1, 1))
        self.fc = nn.Linear(64, num_classes)  # 64: layer3의 출력 채널 수

    #입력 데이터 받아 예측 결과 출력
    def forward(self, x):
        print(f"Input shape: {x.shape}")
        x = self.layer1(x)
        print(f"After layer1: {x.shape}")
        x = self.layer2(x)
        print(f"After layer2: {x.shape}")
        x = self.layer3(x)
        print(f"After layer3: {x.shape}")
        x = self.adaptive_pool(x)
        print(f"After adaptive_pool: {x.shape}")
        x = x.view(x.size(0), -1)
        print(f"After flatten: {x.shape}")
        x = self.fc(x)
        print(f"After fc: {x.shape}")
        #if self.fc is None:
            # Initialize the linear layer dynamically based on the flattened size
        #    self.fc = nn.Linear(x.size(1), self.num_classes).to(x.device)

        #x = self.fc(x)
        return x
```

```python
# AudioDataset 인스턴스 생성
audio_dir_validation = "D:\\대학\\SW\\4학년\\음성인식\\dataset\\004.다화자 가창 데이터\\01.데이터\\2.Validation\\원천데이터"
label_dir_validation = "D:\\대학\\SW\\4학년\\음성인식\\dataset\\004.다화자 가창 데이터\\01.데이터\\2.Validation\\라벨링데이터"
train_dataset = AudioDataset(root_audio_dir=audio_dir, root_label_dir=label_dir)
valid_dataset = AudioDataset(root_audio_dir=audio_dir_validation, root_label_dir=label_dir_validation)

# 데이터 로더 설정
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, collate_fn=collate_fn)
valid_loader = DataLoader(valid_dataset, batch_size=16, shuffle=False, collate_fn=collate_fn)

# 데이터셋에서 첫 번째 아이템 로드
first_data, _ = train_dataset[0]
input_shape = (1, first_data.shape[0], first_data.shape[1])  # 데이터셋 아이템의 차원을 얻음
# 모델 초기화
model = PitchCNN(num_classes=12)#, input_shape=(1, 60, 28786))

# CQT 스펙트럼의 shape 계산
example_idx = 0
example_cqt, _ = train_dataset[example_idx]
input_shape = example_cqt.shape

# 손실 함수와 옵티마이저
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
#데이터 로더에서 배치 처리 어떻게 할건지
def collate_fn(batch):
    inputs, labels = zip(*batch)
    # 최대 길이 계산
    max_length = max(input.shape[1] for input in inputs)  # input.shape[1]는 시간 차원을 나타냄
    # 패딩 적용
    padded_inputs = [F.pad(input, (0, max_length - input.shape[1], 0, 0)) if input.shape[1] < max_length else input for input in inputs]
    # 채널 차원 추가 및 스택
    inputs_tensor = torch.stack([input.unsqueeze(0) for input in padded_inputs])  # 이제 각 입력이 같은 크기를 가져야 함
    # 레이블 텐서 생성
    labels_tensor = torch.tensor(labels, dtype=torch.long)
    return inputs_tensor, labels_tensor


# 모델 평가 로직 검토
model.eval()
with torch.no_grad():
    for inputs, labels in valid_loader:
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        # 출력과 손실을 로깅하여 검토
        print("Outputs:", outputs)
        print("Loss:", loss.item())
```

```python
# 모델 훈련
# param: 훈련 로더, 검증 로더, 손실함수, 옵티마이저, 에폭 수 
def train_model(model, train_loader, valid_loader, criterion, optimizer, num_epochs=25):
    train_losses = []
    valid_losses = []

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        for inputs, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item() * inputs.size(0)
        
        train_loss = running_loss / len(train_loader.dataset)
        train_losses.append(train_loss)

        model.eval()
        running_loss = 0.0
        with torch.no_grad():
            for inputs, labels in valid_loader:
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                running_loss += loss.item() * inputs.size(0)
        
        valid_loss = running_loss / len(valid_loader.dataset)
        valid_losses.append(valid_loss)

        print(f'Epoch {epoch+1}/{num_epochs} - Train Loss: {train_loss:.4f}, Valid Loss: {valid_loss:.4f}')
    
    return train_losses, valid_losses
```

```python
# 손실 검증 시각화
def plot_losses(train_losses, valid_losses):
    plt.figure(figsize=(10, 5))
    plt.plot(train_losses, label='Training Loss')
    plt.plot(valid_losses, label='Validation Loss')
    plt.title('Loss Graph')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python



# 모델 훈련
train_losses, valid_losses = train_model(model, train_loader, valid_loader, criterion, optimizer, num_epochs=10)

# 훈련 및 검증 손실 그래프 그리기
plot_losses(train_losses, valid_losses)
```
